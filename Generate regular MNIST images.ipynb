```python
# Generate random samples from MNIST

import torch
import torchvision
import matplotlib.pyplot as plt
import math

# Load the MNIST dataset
original_dataset = torchvision.datasets.MNIST(root='./data', train=True, download=True)

num_images = 10

def visualize_mnist_samples(dataset, num_samples):
    fig, axes = plt.subplots(2, 5, figsize=(12, 6))
    fig.suptitle('MNIST - Random Samples')

    for i in range(num_samples):
        row = i // 5
        col = i % 5
        image, label = dataset[i]
        
        ax = axes[row, col]
        ax.imshow(image, cmap='gray')
        ax.set_title(f'Label: {label}')
        ax.axis('off')

    plt.tight_layout()
    plt.show()

# Visualize num_images sample images
visualize_mnist_samples(original_dataset, num_images)
```

```python
# Generate one image from each class from MNIST

import torch
import torchvision
import matplotlib.pyplot as plt
import math
import numpy as np

# Load the MNIST dataset
original_dataset = torchvision.datasets.MNIST(root='./data', train=True, download=True)

num_images = 10

def visualize_mnist_samples(dataset, num_samples):
    fig, axes = plt.subplots(2, 5, figsize=(12, 6))
    fig.suptitle('MNIST - Random sample from each class')

    vals_found = []
    num_found = 0
    while num_found < 10:
        j = np.random.randint(0, 60000) # a random number from 0 to 59999
        image, label = dataset[j]
        if label not in vals_found:
            vals_found.append(label)
            row = label // 5
            col = label % 5
            ax = axes[row, col]
            ax.imshow(image, cmap='gray')
            ax.set_title(f'Label: {label}')
            ax.axis('off')
            num_found += 1

    plt.tight_layout()
    plt.show()

# Visualize 10 sample images
visualize_mnist_samples(original_dataset, num_images)
```
